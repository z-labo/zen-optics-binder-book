# Exit angle from a fiber

## Flat surface

출사면이 평평한 파이버 코어를 전파하는 빛을 생각해 보자. 
\
![Fig. 1](.figure/fig1.jpg)

\
전반사 조건을 만족하기 위해서 내부 입사각 $\theta_1$은 임계각 $\theta_c$보다 커야 한다. 이때, 출사 경계면에서의 굴절각 $\phi_1$은 기하적으로 다음과 같이 주어진다.

$$
\phi_1 = \frac{\pi}{2} - \theta_1
$$

경계면에서의 스넬의 법칙 $n_1 \sin \phi_1 = n_0 \sin \phi_0$ 에, 위의 표현을 대입하면 

$$
n_1 \cos \theta_1 = n_0 \sin \phi_0
$$

따라서 
$\phi_0$의 최대값은 $\theta_1 = \theta_c
$
일 때 얻어지며,

$$
\sin \phi_0^{\max} = \frac{n_1}{n_0} \cos \theta_c
$$

임계각 정의$
\sin \theta_c = \frac{n_2}{n_1}$ 을 이용하면

$$
\cos \theta_c = \sqrt{1 - \left(\frac{n_2}{n_2}\right)^2}
$$

따라서 최종적으로

$$
\sin \phi_0^{\max} = \frac{\sqrt{n_1^2 - n_0^2}}{n_0}
$$


이 된다.



## Tapered surface - 1

이번에는 문제를 조금 변형하여, 출사면이 평평하지 않고, 각도 $\beta$ 를 가지는 tapered 단면을 가진다고 가정하자.
![Fig. 2](.figure/fig2.jpg)/

먼저, 코어의 중심을 따라서 직진하는 빛을 고려하면 경계면에서

$$
n_1 \sin\phi_1 = n_0 \sin\phi_0
$$

가 성립된다. $\phi_1 = \dfrac{\pi}{2} - \beta$, $\theta_0 = \phi_0 - \phi_1$ 를 대입하면,

$$
\theta_0 = \sin^{-1}\!\left(\frac{n_1}{n_0}\cos\beta\right) - \left(\frac{\pi}{2} - \beta\right)
$$

를 도출할 수 있다. 다만 $\phi_1$ 은 코어-외부의 경계면에서 critical angle $\theta_c$ 보다 작아야 출사가 이루어지므로, $\beta$ 는 $\dfrac{\pi}{2} - \theta_c$ 보다 커야 한다.

👉 Click the rocket icon (🚀) → Binder

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
n1 = 1.447
n0 = 1.0   # 공기 

# critical angle
theta_c = np.arcsin(n0 / n1)

# cutoff beta (방출 시작점)
beta_cutoff = np.pi/2 - theta_c

# beta 범위 (0 ~ 90도)
beta = np.linspace(0, np.pi/2, 1000)

# 정의역 체크
argument = (n1 / n0) * np.cos(beta)

# 유효 영역 마스크
valid = argument <= 1

# theta0 계산 (invalid는 NaN 처리)
theta0 = np.full_like(beta, np.nan)
theta0[valid] = np.arcsin(argument[valid]) - (np.pi/2 - beta[valid])

# plot
plt.figure()

plt.plot(beta * 180/np.pi, theta0 * 180/np.pi, label=r'$\theta_0(\beta)$')

# cutoff 표시
plt.axvline(beta_cutoff * 180/np.pi, linestyle='--', label='cutoff')

# beta = theta_c (최대점)
plt.axvline(theta_c * 180/np.pi, linestyle=':', label='beta = theta_c')

plt.xlabel(r'$\beta$ (deg)')
plt.ylabel(r'$\theta_0$ (deg)')
plt.title('Emission angle vs taper angle')
plt.legend()
plt.grid()

plt.show()
```

## Tapered surface - 2

Fig. 2에서 좀 더 변형하여, 전반사하면서 출사되는 빛을 생각해 보자.
![Fig. 3](.figure/fig3.jpg)/
마찬가지로 출사경계면에서 $n_1 \sin\phi_1 = n_0 \sin\phi_0$ 가 성립된다.하지만 이번에는 $\phi_1$ 의 값이 
$$
\phi_1 = \frac{\pi}{2} - \beta - (\frac {\pi}{2} - \theta_1) = \theta_1 - \beta
$$
이 되므로 
$$
\theta_0 = \phi_0 - \phi_1
$$
를 아래 식에 대입하면, 
$$
\theta_0 = /sin^{-1}[\frac{n_1}{n_0} /sin(\theta_1 - \beta)] - (\theta_1 - \beta)
$$
를 도출할수있다. 다만 여기서는 2개의 경계면에서의 전반사 조건을 만족해야 한다.
1. $\theta_1$는 코어-클래딩의 경계면에서 critical angle $\theta_{c1}$보다 커야 한다.즉 $\theta_1 \ge sin^{-1}(\frac{n_2}{n_1})$
2. $\phi_1$은 코어-외부의 경계면에서 critical angle $\theta_{c2}$보다 작아야 한다.즉 $\theta_1-\beta<sin^{-1}(\frac{n_0}{n_1})$

이래 코드를 실행해 보면, $\beta$와 출사각 $\theta_0$의 관계를 알수있다. 또한, 현재 조건에서는 출사각의 최대치에 해당하는$\frac{\partial \theta_0}{\partial \beta}=0$지점에서  $\beta$는 cutoff angle보다 작아서 사실상 cunoff angle에서의 출사각이 최대치가 되는것을 확인할수있다.

👉 Click the rocket icon (🚀) → Binder

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# refractive indices
n1 = 1.447
n2 = 1.431
n0 = 1.0

# critical angles
theta_c1 = np.arcsin(n2 / n1)
theta_c2 = np.arcsin(n0 / n1)

# parameter ranges
theta1 = np.linspace(theta_c1 * 1.01, np.pi/2 - 1e-3, 200)
beta = np.linspace(0, np.pi/2, 200)

THETA1, BETA = np.meshgrid(theta1, beta)

# phi1
phi1 = THETA1 - BETA

# 조건 1: core-cladding TIR
cond1 = THETA1 > theta_c1

# 조건 2: 출사면에서 TIR 방지
cond2 = phi1 < theta_c2

# Snell 정의역
argument = (n1 / n0) * np.sin(phi1)
cond3 = argument <= 1

# 전체 조건
valid = cond1 & cond2 & cond3

# theta0 계산
theta0 = np.full_like(THETA1, np.nan)
theta0[valid] = np.arcsin(argument[valid]) - phi1[valid]

# 최대값 찾기
max_idx = np.nanargmax(theta0)
max_theta0 = np.nanmax(theta0)

beta_max = BETA.flatten()[max_idx]
theta1_max = THETA1.flatten()[max_idx]

# plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(
    BETA * 180/np.pi,
    THETA1 * 180/np.pi,
    theta0 * 180/np.pi,
    cmap='viridis'
)

# 최대점 표시
ax.scatter(
    beta_max * 180/np.pi,
    theta1_max * 180/np.pi,
    max_theta0 * 180/np.pi,
    color='red',
    s=50,
    label='Max'
)

ax.set_xlabel(r'$\beta$ (deg)')
ax.set_ylabel(r'$\theta_1$ (deg)')
ax.set_zlabel(r'$\theta_0$ (deg)')
ax.set_title('3D map of emission angle')

ax.legend()
plt.show()

# 최대값 출력
print("Max theta0 (deg):", max_theta0 * 180/np.pi)
print("At beta (deg):", beta_max * 180/np.pi)
print("At theta1 (deg):", theta1_max * 180/np.pi)
```
