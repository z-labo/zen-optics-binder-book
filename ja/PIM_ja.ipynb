## STEP 1 — 伝搬定数 β とは何か？

光ファイバーの中では、光は**モード**(mode) と呼ばれる特定の空間パターンで伝搬する。  
各モードは固有の**伝搬定数 $\beta$**を持ち、これは軸方向（z 方向、または縦方向）に距離 $z$ だけ進んだとき、位相が $e^{i\beta z}$ だけ付加されることを意味する。

### β の物理的意味

自由空間での波数は $k_0 = \omega/c$ である。  
屈折率 $n$ の一様媒質中では $n k_0$ になる。  
しかし光ファイバーは、コア（屈折率 $n_\text{core}$）とクラッド（屈折率 $n_\text{clad}$、$n_\text{clad} < n_\text{core}$）からなる**導波構造**であるため、光はその中間の有効屈折率で伝搬しているかのように振る舞う。

$$
n_\text{clad} k_0 < \beta < n_\text{core} k_0
$$

この範囲の外ではモードは存在しない。

- $\beta \geq n_\text{core} k_0$ の場合、コア全体の位相が同じになってしまい、伝搬解は存在しない。
- $\beta \leq n_\text{clad} k_0$ の場合、クラッド側で減衰せず、エネルギーが無限に漏れ出す**放射モード**(radiation mode) になる。

また STEP 7 で説明するように、両側の境界条件を同時に満たす $\beta$ は離散的にしか存在せず、それによって構成される横方向電磁場の分布が**導波モード**(guided mode) である。

---

## STEP 2 — 横方向波数の定義

伝搬定数 $\beta$ が決まると、コア内とクラッド内での**横方向**（x-y 平面内の振動）の振る舞いを記述する二つの波数を定義できる。

### コア内部の横方向波数 $\kappa$

コアの中では、光は横断面方向に**振動**(oscillation) する。

$$
\kappa = \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$

$k_0 n_\text{core}$ はコア内での全波数であり、そこから軸方向成分 $\beta$ を差し引くと、残りが横断面方向成分になる（ピタゴラス的なベクトル分解）。$\kappa$ が実数になるには $\beta < k_0 n_\text{core}$ である必要があり、これは前述の上限条件と一致する。

### クラッド内部の減衰定数 $\sigma$

クラッドの中では、光は振動する代わりに**指数関数的に減衰**(evanescent) する。

$$
\sigma = \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

$\sigma$ が実数になるには $\beta > k_0 n_\text{clad}$ である必要があり、これは下限条件と一致する。  
この二つの条件をともに満たす $\beta$ の範囲でのみ、完全な導波モードが存在する。

### 複素屈折率 (Complex refractive index) $N = n + ik$

平面波の波動関数を $\psi(x) = Ae^{ikx}$ と書くと、$k = Nk_0$ であるため、

$$
\psi(x) = Ae^{i(k_0n+ik_0k)x} = Ae^{-k_0kx}e^{ik_0nx}
$$

すなわち、$e^{-k_0kx}$ は $x$ に応じて振幅が減衰定数 $k_0k$ で急速に減衰する**減衰項**、$e^{ik_0nx}$ は波数 $k_0n$ で損失なく振動する**振動項**と見なすことができる。これが上で述べたコア内部の横方向波数 $\kappa$ とクラッド内部の横方向減衰定数 $\sigma$ に数学的にどのように整合するかは、後で議論する。

## STEP 3 — 規格化横方向波数 $u$、$w$

数式を扱いやすくするため、コア半径 $a$ で規格化した無次元波数を導入する。

$$
u = \kappa a = a \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$

$$
w = \sigma a = a \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

この二つの量は次の関係を満たす。

$$
u^2 + w^2 = a^2 k_0^2 (n_{\mathrm{core}}^2 - n_{\mathrm{clad}}^2) \equiv V^2
$$

ここで $V$ は**開口数**(numerical aperture) とコア半径を合わせた**V パラメーター**または**V 数**であり、光ファイバーのモード数を決定する重要な量である。  
$V < 2.405$ の場合、基本モード $HE_{11}$ のみが伝搬可能な**シングルモードファイバー**となる。

---

## STEP 4 — Bessel 関数の導関数表記

境界条件を書くとき、Bessel 関数 $J_l(r)$、$K_l(r)$ の導関数が頻繁に現れる。  
これを簡潔に表記するため、次の記号を導入する。

1. **第一種 Bessel 関数の微分 $J_l'(r)$**
   * $J_d(l, r) = J_{l-1}(r) - \frac{l}{r} J_l(r)$
   * この定義は第一種 Bessel 関数の微分公式と同じである。  
   * シミュレーションでは $J_d(l, r) = J_l'(r)$ とする。

2. **第二種変形 Bessel 関数の微分 $K_l'(r)$**
   * $K_d(l, r) = \frac{l}{r} K_l(r) - K_{l+1}(r)$
   * この定義は第二種変形 Bessel 関数の微分公式と同じである。  
   * シミュレーションでは $K_d(l, r) = K_l'(r)$ とする。

この表記を導入すると、超越方程式中の導関数を $J_l$、$K_l$ および次数が一つ異なる $J_{l\pm1}$、$K_{l\pm1}$ の**線形結合**(Linear Combination) として簡潔に変換できるため、数値計算およびコード実装が非常に便利になる。

## STEP 5 — 非対称パラメーター $n_d$、$n_s$

コアとクラッドの屈折率差を、二つの組み合わせで定義する。

$$
n_d = \frac{n_{\mathrm{core}}^2 - n_{\mathrm{clad}}^2}{2\,n_{\mathrm{core}}^2}
$$

$$
n_s = \frac{n_{\mathrm{core}}^2 + n_{\mathrm{clad}}^2}{2\,n_{\mathrm{core}}^2}
$$

- $n_d$ はコアとクラッドの屈折率の**差**を表す非対称パラメーターであり、境界条件において電場と磁場の接線成分がどの程度異なって結合するかを反映する。  
  弱導波近似 ($n_\text{core} \approx n_\text{clad}$) では $n_d \approx 0$ となり、式が単純化される。

- $n_s$ は屈折率の**平均**に近い値であり、境界での境界条件のスケールを調整する役割を持つ。

---

## STEP 6 — 補正項 $R(\beta)$

$R(\beta)$ はベクトル波動方程式から生じる**偏光補正項**であり、EH モードと HE モードを区別する核心的な役割を果たす。

$$
R(\beta) = \sqrt{
  n_d^2 \left[\frac{K_l'(w)}{w\,K_l(w)}\right]^2
  + \left(\frac{l\,\beta}{n_{\mathrm{core}}\,k_0}\right)^2
  \left(\frac{1}{w^2} + \frac{1}{u^2}\right)^2
}
$$

- **第一項** $n_d^2[\cdots]^2$: 屈折率の不連続による偏光混合効果を表す。$n_d$ が大きいほど（屈折率差が大きいほど）、この項は大きくなる。

- **第二項**: 方位角次数 $l$ と伝搬定数 $\beta$ に依存する。$l = 0$ の場合、この項は消え、TE/TM モードだけが残る。

$R(\beta)$ は常に正の実数であり、この値の**符号**を変えて取ることで、EH モードと HE モードの超越方程式を統一的に記述できる。

---

## STEP 7 — 超越方程式と β の決定

### EH モードの超越方程式

$$
f_{\mathrm{EH}}(\beta) = \frac{J_{l+1}(u)}{u\,J_l(u)} - n_s\,\frac{K_l'(w)}{w\,K_l(w)} - \left(\frac{l}{u^2} - R(\beta)\right) = 0
$$

### HE モードの超越方程式

$$
f_{\mathrm{HE}}(\beta) = \frac{J_{l+1}(u)}{u\,J_l(u)} - n_s\,\frac{K_l'(w)}{w\,K_l(w)} - \left(\frac{l}{u^2} + R(\beta)\right) = 0
$$

### 二つの方程式の違い

二つの超越方程式の**唯一の違い**は、最後の項における $R(\beta)$ の符号である。

| モード | 最後の項 | 意味 |
|--------|----------|------|
| EH     | $\frac{l}{u^2} - R(\beta)$ | $R(\beta)$ を**引く**方向への偏光補正 |
| HE     | $\frac{l}{u^2} + R(\beta)$ | $R(\beta)$ を**足す**方向への偏光補正 |

この符号の違いは、境界条件において電場の横方向成分がコアとクラッドの境界でどのように接続されるかに由来し、**電場が支配的なモード (EH)** と**磁場が支配的なモード (HE)** の構造差を反映している。

### β の数値的決定方法

1. 各 $l$ について、$\beta \in (n_\text{clad} k_0,\; n_\text{core} k_0)$ の範囲を離散的に探索する。
2. $f_{\mathrm{EH}}(\beta)$ または $f_{\mathrm{HE}}(\beta)$ が 0 になる点を**符号変化**(sign change) によって見つける。
3. 二分法 (bisection) または Brent 法などで精密に収束させる。
4. 各解 $\beta_{lm}$（方位角次数 $l$、その方位角次数に対する複数の解をインデックス付けした半径方向次数 $m$）に対応するモードプロファイルを構成する。

## STEP 8 — モード命名規則

| モード種類 | 説明 | 存在条件 |
|------------|------|----------|
| $TE_{0m}$ | 電場が横断面内にのみ存在、$l=0$ | EH 方程式 $l=0$ の特殊解 |
| $TM_{0m}$ | 磁場が横断面内にのみ存在、$l=0$ | HE 方程式 $l=0$ の特殊解 |
| $HE_{lm}$ | 磁場成分が優勢な混合モード | HE 方程式 $l \geq 1$ |
| $EH_{lm}$ | 電場成分が優勢な混合モード | EH 方程式 $l \geq 1$ |

基本モード $HE_{11}$ は $l=1$ の HE 方程式で $m=1$ に対応する最初の解であり、**遮断周波数が 0** であるため、どの V 数でも常に存在する。

**TE モード** ベクトル波動方程式で $E_z = 0$ となるモード<br>
**TM モード** ベクトル波動方程式で $H_z = 0$ となるモード

## STEP 9 — 理想状態における $\beta_{lm}$ に対する各モードの空間分布を求める

円筒形誘電体光ファイバー (Optical Fiber) 内部を伝搬する**ハイブリッドモードの電磁場**を、円筒座標系 ($\rho, \theta, z$) で計算する式として実装する。

### モード別振幅係数の求め方 ($A, B, C, D$)
光ファイバー電磁場の z 軸成分の振幅を決定する定数である。$l$ の値（方位角次数）に応じて式が変わる。

1. $l = 0$ の場合
    * **EH mode** (TM-like)
      $\displaystyle A = 1, B = 0, C = \frac{J_0(u)}{K_0(w)}, D = 0$
    * **HE mode** (TE-like)
      $\displaystyle A = 0, B = 1, C = 0, D = \frac{J_0(u)}{K_0(w)}$

2. $l \neq 0$ の場合（ハイブリッドモード）
    * $A = 1$
    * $B = \frac{il\beta}{\omega\mu_0} \left( \frac{1}{(w)^2} + \frac{1}{(u)^2} \right) \left[ \frac{J_l'(u)}{u J_l(u)} + \frac{K_l'(w)}{w K_l(w)} \right]^{-1}$
    * $C = \frac{J_l(u)}{K_l(w)}A$
    * $D = B \cdot C$

### 領域別電磁場の式
* **共通位相項 (Phase term):** $\Phi = e^{i(l\theta - \beta z)}$
* 実際のシミュレーションでは、コア領域での $E_r, E_\theta$ のみを使用する。

1. コア領域 ($\rho < a$)：導波（振動）成分

$$
\begin{aligned}

E_r &= -\frac{i\beta}{\kappa^2} \left[ A \kappa J_l'(\kappa\rho) + \frac{i\omega\mu_0 l}{\beta\rho} B J_l(\kappa\rho) \right] \Phi \\

E_\theta &= -\frac{i\beta}{\kappa^2} \left[ \frac{il}{\rho} A J_l(\kappa\rho) - \frac{\omega\mu_0}{\beta} B \kappa J_l'(\kappa\rho) \right] \Phi \\

E_z &= A J_l(\kappa\rho) \Phi \\

\\

H_r &= -\frac{i\beta}{\kappa^2} \left[ B \kappa J_l'(\kappa\rho) - \frac{i\omega\epsilon_{\text{core}} l}{\beta\rho} A J_l(\kappa\rho) \right] \Phi \\

H_\theta &= -\frac{i\beta}{\kappa^2} \left[ \frac{il}{\rho} B J_l(\kappa\rho) + \frac{\omega\epsilon_{\text{core}}}{\beta} A \kappa J_l'(\kappa\rho) \right] \Phi \\

H_z &= B J_l(\kappa\rho) \Phi \\

\\

\end{aligned}
$$

2. クラッド領域 ($\rho > a$)：減衰（エバネッセント）成分

$$
\begin{aligned}

E_r &= \frac{i\beta}{\sigma^2} \left[ C \sigma K_l'(\sigma\rho) + \frac{i\omega\mu_0 l}{\beta\rho} D K_l(\sigma\rho) \right] \Phi \\

E_\theta &= \frac{i\beta}{\sigma^2} \left[ \frac{il}{\rho} C K_l(\sigma\rho) - \frac{\omega\mu_0}{\beta} D \sigma K_l'(\sigma\rho) \right] \Phi \\

E_z &= C K_l(\sigma\rho) \Phi \\

\\

H_r &= \frac{i\beta}{\sigma^2} \left[ D \sigma K_l'(\sigma\rho) - \frac{i\omega\epsilon_{\text{clad}} l}{\beta\rho} C K_l(\sigma\rho) \right] \Phi \\

H_\theta &= \frac{i\beta}{\sigma^2} \left[ \frac{il}{\rho} D K_l(\sigma\rho) + \frac{\omega\epsilon_{\text{clad}}}{\beta} C \sigma K_l'(\sigma\rho) \right] \Phi \\

H_z &= D K_l(\sigma\rho) \Phi \\

\\

\end{aligned}
$$
