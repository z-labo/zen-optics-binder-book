# Propagation indipendent modes (PIM)

## CONDITION

코어반경이 D=25um, 파장 \lambda=633nm, NA=0.22인 광학 파이버를 상상해보자.  

## STEP 1
먼자 전파상수 $\beta$ 를 상상해보자. 특정 PIM모드에 대해서 파이버내에서 거리 $z$ 를 전파했을때, 출력면에서 모드의 위상은 $e^{i \beta z}$ 만큼 추가된다. \
$\beta$는 $n_{clad}< \beta < n_{core}$ 내에 이산적으로 존재할수있다. 그러면 $\beta$는 어떻게 구하고, 그에 대응하는 PIM모드는 무엇인지 알아보자.

먼저 コア内の横方向波数 $\kappa$ とクラッド側の減衰定数 $\sigma$ 는 이하의 식으로 표현된다.
$$
\kappa = \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$
$$
\sigma = \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

그러면 normalized transverse wave numbers are
$$
u = \kappa a = a \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$
$$
w = \sigma a = a \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

또한 $K_d(l,r) = \frac{l}{r} K_l(r) - K_{l+1}(r)$ 로 적으면, 
$$
K_d(l,r) = K'_l(r)
$$
이 성사된다. 그중$K_l(r)$ 는 modified bessel function of the second kind

또한 
$$
n_d = \frac{n_{core}^2 - n_{clad}^2}{2 n_{core}^2}
$$

$$
n_s = \frac{n_{core}^2 + n_{clad}^2}{2 n_{core}^2}
$$

Set
$$
R(\beta) = \sqrt{n_d^2[\frac{k_l'(w)}{w k_l(w)}]^2 + (\frac {l \beta}{n_{core} k_0})^2(\frac{1}{w^2} + \frac{1}{u^2})}
$$
EH모드의 초월 방정식
$$
f_{EH} = \frac{J_{l+1}(u)} {u J_l(u)} - n_s \frac {K_l'(w)} {w K_l(w)} - (\frac{l}{u^2}-R(\beta))
$$

$f_{EH} = 0 $ 을 만족하는 $\beta$ 를 찾으면 되는것. 마찬가지로 HE모드의 초월방정식을 쓰면 
$$
f_{HE} = \frac{J_{l+1}(u)} {u J_l(u)} - n_s \frac {K_l'(w)} {w K_l(w)} - (\frac{l}{u^2}-R(\beta))
$$
에서 $f_{HE} = 0 $을 만족하는 $\beta$ 를 찾으면 되는데, 이때 두개의 초월방정식의 다른점은 마지막 $R(\beta)$의 부호에 달려있음을 알수있다


## STEP 2
\
$l$의 최대치를 설정하고 (예를들면 $L = l_{MAX} = 100$) 0부터 차례로 대입해서 $\beta$ 를 찾는다. $l$은 OAM의 인덱스를 나타내며, 각도방향으로 위상이 몇사이클 변하는지를 나타낸다. 또한 위상이 변하는 방향은 시계방향, 시계반대방향 두개 상태가 존재할수있으며, $\pm l$로 나타낸다.
때로는 같은 $l$에 대하여 복수의 $\beta$ 를 구할수있는데, $m = 1,2,3...$으로 인덱싱한다. 이것을 Radical index라고 부른다.
