## 4．本テーマにおけるTA業務について
### 4.1 業務内容

#### 授業開始前：授業開始10分前を目安に情報工学実験室に集合
* ドアの開放，バリアフリー用のスロープの設置（電動車椅子使用の要配慮学生がいるCグループの回（5/7, 5/12, 5/14）のみ）
* 機材の収納されている段ボールを机の上に出す
* 機材を各班の机に配布．
* PC本体（対応する班番号），ディスプレイ，キーボード，マウス，電源ケーブル2本（PC，ディスプレイ用），LANケーブル2本（パケット観測装置あり・なし各1本）
* サーバPCの設置
* スイッチングハブの設置とサーバ周りのネットワークケーブルの配線
* サーバからのLANケーブルは，ネットワークハブの8番に接続

#### 授業中： [担当範囲](https://z-labo.github.io/zen-optics-binder-book/jb2026.html#id11)の学生を中心に指導補助
* 出席チェック
* 遅刻した学生については出席表に分単位で遅刻時間を記録してください
* 質問対応の他、実施状況を主体的に確認し、臨場感を持って活発に参加できるように積極的に声をかけてあげてください
* 受講生の学習を促すように、どのように考えるのか、指針を示し導くような指導補助をお願いします
* 機材の取り扱いについて、正しくない使用方法をとっているときには正しい使用方法を教えてあげてください
* 促しているにも関わらず実験に参加している様子がない，席を外して他の人と無駄話をしていてサボっている等，受講態度に問題がある学生については，出席表にメモしておいてください

#### 授業終了後
* 機材の片付け、段ボールを棚に戻す。
* メールで当日中に [報告書](https://z-labo.github.io/zen-optics-binder-book/jb2026.html#id12)を提出

#### ラウンド終了時（各グループ終了時、計3回）
* 各班PCの[設定初期化作業](https://z-labo.github.io/zen-optics-binder-book/jb2026.html#id14)
* 初回提出レポートの[書式チェック](https://z-labo.github.io/zen-optics-binder-book/jb2026.html#id15)

#### 混雑時の対応(3日目のみ)
* クライアント-サーバ間のパケット伝送に関して，サーバ/ルータへのアクセス集中を予防するために， [実験状況確認シート](https://docs.google.com/spreadsheets/d/1mPgSMAgqTgaBmLnlQG-bsSq9ibBL3UDG7k5OipLJPNY/edit?usp=sharing) の共同編集と参照による各班の順次実行の運用サポートをおねがいします。
* スマホなどで[実験状況確認シート](https://docs.google.com/spreadsheets/d/1mPgSMAgqTgaBmLnlQG-bsSq9ibBL3UDG7k5OipLJPNY/edit?usp=sharing) のリアルタイム編集・参照ができるように準備をおねがいします。
* パケット伝送（ping，send/recv）を実行するのは同時刻において一班のみとします。そのため1班から7班まで実行の順番を回していきます。

   **手順**
   1. 【受講生】順番が回ってくる前に，次の計測で実施するコマンドを入力した状態で待機
   2. 【受講生】待機中は，計測済みのデータ整理，考察を実施
   3. 【TA】自身の担当班に順番が回ってきたら，該当のステータスを「実行中（■）」に修正し，受講生に実行の指示．原則1回の順番で1回の計測．2~3回は許容するが，それ以上の実行は3.として取り扱う
   4. 【TA】2分以内を目安に実行ができない場合は，この回の順番は終了．該当のステータスを「NG（×）」に修正．次の班担当のTAに順番を回すことを伝達
      * 順番が回って来てから送信コマンドのパラメータを考えるようなケースが該当
   5. 【TA】実行ができた場合，該当のステータスを「OK（○）」に修正．次の班担当のTAに順番を回すことを伝達

### 4.5 レポートのチェック
#### 4.5.1 チェック方法
* メールで共有される担当の割り当てファイルに従って，担当シートの該当の項目の書式チェック
   * ファイル名中の*は，グループを表すA/B/C
   * チェック方法については「チェック方法」シートに記載
   * 作業量の削減のため一項目一人でのチェックになるため，それを念頭に丁寧なチェックをお願いします
* 担当項目について
   * メールを参照
   * 慣れもあるため、3グループのチェックすべてで統一
   * ただし，負担の平均化のために要望があれば，皆さんで相談して交換してもらっても構いません
   * 体調不良や就職活動、論文投稿等の理由により余裕がない場合はTA内で代役等の調整をお願いします
