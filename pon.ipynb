# XGS-PON (10-Gigabit Symbetric Passive Optical Network) 

## 次世代光ファイバー通信規格（ITU-TG.9807.1）
\
家庭用・法人用高速インターネット: NURO光などの高速通信サービスで導入が進んでいる。
XGS-PON의 대표 파장은 downstream 1577 nm근처, upstream 1270 nm근처로 알려져 있다
Dense Wavelength Division Multiplexing(DWDM)은 다양한 광파장을 사용하여 여러 데이터 채널을 가진 단일 광섬유 멤버를 통해 데이터를 전송하는 가장 진보된 방법 중 하나입니다. 광섬유 입력의 용량 밀도와 다양성을 개선하여 빠른 데이터 부스트를 가능하게 하며, 이는 현대 통신의 기반을 형성합니다. 네트워크 및 데이터 센터. 각 채널은 별도의 파장에 있으므로 서로 영향을 미치지 않고 동시에 방대한 양의 정보를 전송할 수 있습니다. 이는 설정된 거리에서 빠르고 안정적인 통신을 위한 필수적인 기반을 형성하여 인프라를 늘리고 필요한 추가 파이버를 줄입니다.
DWDM 시스템에서 가장 기본적인 장치는 광 트랜스폰더, 멀티플렉서, 디멀티플렉서, 광 증폭기입니다. 트랜스폰더를 사용하면 전기 신호가 적절한 파장의 광 신호로 변환되어 DWDM이 작동할 수 있습니다. 멀티플렉서는 다양한 파장의 수많은 신호가 하나의 파이버로 결합되어 데이터 대역폭을 효과적으로 사용하는 광 통신에 사용됩니다. 디멀티플렉서는 반대로 결합된 광 신호를 광섬유에서 분리하여 추가 처리를 위한 구성 요소별 신호로 만듭니다. 에르븀 도핑 파이버 증폭기(EDFA)는 신호를 전기 형태로 다시 변환하지 않고 장거리에서 신호 강도를 높이는 광 증폭기입니다. 이러한 요소는 광 시스템에서 효과적인 광대역 데이터 전송을 위한 충분한 대용량 채널을 보장합니다.
DWDM멀티플렉서와 디멀티플렉서는 주로 AWG(Arrayed Waveguide Grating) 기술을 사용.
광 트랜스폰더는 광섬유와 전기신호와의 쌍방향 변환을 실시. 예를 들어, O(1310)-E-O(1550)광 전환 유닛일때 파장을 직접 바꾸기가 어려우므로, 파장변환을 해주는 트랜스폰더를 사용하는것이다. 트랜스폰더는 트랜시버의 광전변환 기능에 더하여, 파장변환 기능을 추가로 가진다. 송신단, 전송, 수신단으로 구성. 또한 각각방향에 파장변환장치 Mux/ Demux가 있음.

## 예시경로

예시 경로:

PC/스마트폰 → 집 공유기 → ONT/ONU → 광분배기, splitter → 통신사 OLT → BNG/라우터 → 메트로/코어 DWDM 광전송망 → IX/CDN/데이터센터 라우터 → 서버

이 중 “진짜 광통신”이 일어나는 구간은 크게 두 부분이다.

가입자망/access 구간

집 ONT/ONU ↔ 통신사 국사 OLT
예: XGS-PON
하향 1577 nm, 상향 1270 nm 근처
세기변조/직접검출, IM-DD
수동 광분배기 사용
메트로/코어/장거리 구간

통신사 라우터/전송장비 ↔ 다른 국사/데이터센터
예: 100G/400G coherent DWDM
C-band, 대략 1530~1565 nm, 즉 약 191~196 THz
DP-QPSK, DP-16QAM 등 coherent 변조
EDFA/Raman 증폭기, ROADM 사용

## 계층구조 
### 1계층 구성
캐리어 주파수, 변조, 전송, 검출, 증폭, 분기가 일어나는 곳

PC/공유기 ─ Ethernet ─ ONT/ONU ────── single-mode fiber ────── splitter ────── OLT
                                      1:N passive optical splitter


ONT/ONU: 집에 있는 광모뎀.
OLT: 통신사 국사에 있는 PON 집선 장비.
ODN: Optical Distribution Network. 중간의 광섬유, 커넥터, 분배기 포함.
splitter: 전원 없는 수동 소자. 예를 들어 1:16, 1:32, 1:64 등으로 한 OLT 포트를 여러 가입자에게 나눈다.

대표값:

방향	파장	대략적 광 캐리어 주파수
OLT → ONT 하향	1577 nm	약 190.1 THz
ONT → OLT 상향	1270 nm	약 236.1 THz

#### 광원

하향, OLT 송신기
보통 DFB laser 또는 EML, Electro-Absorption Modulated Laser 계열을 사용한다.
1577 nm 근처의 좁은 스펙트럼 광원을 만들고, 이 빛의 세기를 데이터에 따라 켜고 끄거나 높고 낮게 바꾼다.

상향, ONT 송신기
ONT는 1270 nm 근처의 burst-mode 레이저 송신기를 쓴다.
중요한 점은 상향은 여러 ONT가 같은 파장을 공유하기 때문에, ONT가 계속 빛을 내면 안 된다.
OLT가 “너는 이 시간 슬롯에 보내라”고 허가한 순간에만 짧게 burst로 송신한다.

#### 변조방식
XGS-PON access 구간은 일반적으로 IM-DD, Intensity Modulation Direct Detection 방식이다.

Intensity Modulation: 빛의 세기를 데이터에 따라 바꿈.
Direct Detection: 수신기가 광전류의 크기를 바로 검출함.실질적으로는 NRZ OOK/PAM-2 계열로 볼 수 있다.
OOK: On-Off Keying.
1이면 빛 강함, 0이면 빛 약함/꺼짐.또는 두 레벨의 광세기.
PON 계열은 오랫동안 IM-DD 기반 NRZ/OOK를 사용해 왔고, XGS-PON도 coherent가 아니라 저가의 직접검출 구조를 사용한다. 

QSFP-DD 400G-ZR/ZR+とは？
これまでトランスポンダーという高価な装置で実現していた、高速大容量（400G x 64＝25.6T）＆長距離（120km～）の光伝送を小さな光
トランシーバーで実現しちゃおう！という野心的な規格
トランスポンダーが不要になり、スイッチ/ルーターとトランスポンダーをつなぐための大量の光トランシーバーも不要になるので、システム全体の劇的な低コスト化、省電力化、省スペース化が可能


#### 전송 방식: 하향 broadcast, 상향 TDMA
하향: OLT → 모든 ONT에 broadcast
OLT가 1577 nm 빛으로 프레임을 보낸다.
OLT ──1577 nm downstream── splitter ── 모든 ONT
splitter는 전기적 스위치가 아니다. 그냥 빛을 나눠 보낸다. 따라서 하향 광신호는 같은 splitter에 붙은 모든 ONT에 도달한다.

하지만 각 ONT는 프레임 안의 식별자, 예를 들어 XGEM Port-ID, 암호화 키 등을 보고 자기에게 온 데이터만 수용한다.

즉 물리적으로는 broadcast, 논리적으로는 가입자별 분리다.

상향: 여러 ONT → OLT, TDMA
상향은 문제가 있다. 여러 ONT가 동시에 1270 nm로 쏘면 충돌한다.

그래서 XGS-PON은 TDMA, Time Division Multiple Access를 쓴다.

Copy시간축:
[ONT A 송신][guard][ONT B 송신][guard][ONT C 송신][guard]...
OLT가 각 ONT에게 upstream bandwidth map/grant를 내려준다.

“ONT A는 t1~t2에 보내라.”
“ONT B는 t3~t4에 보내라.”
“ONT C는 t5~t6에 보내라.”
이 과정을 DBA, Dynamic Bandwidth Allocation가 조절한다. 트래픽이 많은 ONT에게 더 많은 upstream slot을 줄 수 있다.

ranging: 거리 보정
각 ONT는 OLT로부터 거리가 다르다.

A 가입자: 1 km
B 가입자: 8 km
C 가입자: 15 km
그러면 같은 시각에 쏜 빛도 OLT에 도착하는 시간이 다르다. 그래서 OLT는 초기 등록 과정에서 각 ONT의 지연을 측정하고, ONT별 송신 타이밍을 조정한다. 이것이 ranging이다.

결과적으로 OLT 입장에서는 각 ONT burst가 정해진 슬롯에 맞춰 들어온다.

#### 검출방식
ONT 수신기, 하향
ONT 안에는 보통 다음 블록이 있다.

photodiode/APD → TIA → limiting amplifier/equalizer → CDR → bit recovery

photodiode 또는 APD: 빛을 전류로 바꿈.
TIA, transimpedance amplifier: 작은 광전류를 전압 신호로 증폭.
equalizer: 광섬유 분산, 전기적 대역 제한으로 찌그러진 파형 보정.
CDR, Clock and Data Recovery: 클록과 데이터를 복원.
이후 FEC 복호, 프레임 동기, XGEM 처리로 넘어간다.

OLT 수신기, 상향
OLT 쪽 상향 수신기는 더 까다롭다. 왜냐하면 ONT마다 거리가 다르고, 광파워도 다르며, burst가 짧게 들어오기 때문이다.

그래서 burst-mode receiver가 필요하다.

burst마다 빠르게 이득 조정.
decision threshold 조정.
클록 재동기.
각 ONT의 수신 파워 차이 보정.

#### 중간 분기점, splitter에서 일어나는 일
PON의 splitter는 보통 수동 광분배기다.

하는 일:

하향: OLT에서 온 빛을 1:N으로 나눈다.
상향: 여러 ONT에서 온 빛을 N:1로 합친다.
하지 않는 일:

패킷 해석 안 함.
MAC 주소 안 봄.
IP 주소 안 봄.
전기 변환 안 함.
증폭 안 함.
라우팅 안 함.

#### PON 구간에서 증폭은?
일반 FTTH PON access 구간에서는 보통 중간 광증폭기를 두지 않는다.

이름 그대로 Passive Optical Network이기 때문이다.

대신 설계할 때 optical budget을 맞춘다.

광섬유 손실: 1550 nm 근처에서 대략 0.2 dB/km 수준.
20 km면 약 4 dB.
1:32 splitter 손실 약 15~17 dB.
커넥터, 접속, 여유 마진 포함.
총 손실이 OLT/ONT의 optical budget 안에 들어오도록 설계.
XGS-PON 장비는 보통 20 km급 ODN을 많이 전제로 하고, 광 budget class에 따라 허용 손실이 다르다.

#### 물리층에서 데이터 링크층으로 넘기기 전 처리
OLT나 ONT가 광신호를 받은 뒤 바로 Ethernet frame이 나오는 게 아니다. 대략 이런 처리가 있다.

→ 광 검출
→ 전기 신호
→ 증폭/등화
→ 클록 데이터 복원, CDR
→ 비트 스트림 복원
→ 프레임 동기
→ scrambling/descrambling
→ FEC 복호
→ XGTC/XGEM 프레임 해석
→ Ethernet frame 재조립
→ 데이터 링크층으로 전달

XG-PON/XGS-PON 계열은 XGEM encapsulation, fragmentation, delineation 등을 사용한다. ITU-T G.987.3 계열에서 XGEM이 SDU fragmentation, encapsulation, delineation을 지원한다고 설명한다. ITU-T G.987.3

또한 XGS-PON은 RS FEC, Reed-Solomon FEC를 사용한다. 예를 들어 ITU 문서 검색 결과에 따르면 XGS-PON downstream FEC code로 **RS(248,216)**가 언급된다. ITU-T G.9807.1 e-publication 검색 결과
