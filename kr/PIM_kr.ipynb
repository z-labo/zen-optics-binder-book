## STEP 1 — 전파 상수 β란 무엇인가?

광섬유 안에서 빛은 **모드**(mode) 라는 특정한 공간 패턴으로 전파된다.  
각 모드는 고유한 **전파상수 $\beta$** 를 가지며, 이는 축 방향(z 방향, 혹은 종 방향이라고도 부름)으로 거리 $z$ 만큼 진행했을 때 위상이 $e^{i\beta z}$ 만큼 추가됨을 의미한다.

### β의 물리적 의미

자유 공간에서 파수는 $k_0 = \omega/c$ 이다.  
굴절률 $n$ 인 균질 매질 안에서라면 $n k_0$ 가 된다.  
그런데 광섬유는 코어(굴절률 $n_\text{core}$)와 클래드(굴절률 $n_\text{clad}$, $n_\text{clad} < n_\text{core}$)로 이루어진 **도파구조**이므로, 빛은 그 중간 유효 굴절률로 전파하는 것처럼 행동한다.

$$
n_\text{clad} k_0 < \beta < n_\text{core} k_0
$$

이 범위 밖이면 모드는 존재하지 않는다.

- $\beta \geq n_\text{core} k_0$ 이면 코어 전체가 위상이 같아 버려 전파 해가 없다.
- $\beta \leq n_\text{clad} k_0$ 이면 클래드 쪽이 감쇠하지 않아 에너지가 무한히 빠져나가는 **방사 모드**(radiation mode) 가 된다.

또한 STEP 7 에서 설명하겠지만, 양쪽 경계 조건을 동시에 만족하는 $\beta$ 는 이산적으로만 존재하고, 이것이 구성하는 횡방향 전자기장의 분포가 **전파 모드**(guided mode) 다.

---

## STEP 2 — 횡방향 파수의 정의

전파 상수 $\beta$ 가 정해지면, 코어 안과 클래드 안에서의 **횡방향**(x-y평면에서의 진동) 거동을 기술하는 두 가지 파수를 정의할 수 있다.

### 코어 내부의 횡방향 파수 $\kappa$

코어 안에서 빛은 횡단면 방향으로 **진동**(oscillation) 한다.

$$
\kappa = \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$

$k_0 n_\text{core}$ 는 코어 안에서의 전체 파수이고, 그것에서 축 방향 성분 $\beta$ 를 빼면 나머지가 횡단면 방향 성분이다(피타고라스적으로 벡터 분해). $\kappa$ 가 실수가 되려면 $\beta < k_0 n_\text{core}$ 이어야 하며, 이는 앞서 나온 상한 조건과 일치한다.

### 클래드 내부의 감쇠 상수 $\sigma$

클래드 안에서 빛은 진동하는 대신 **지수 함수적으로 감쇠**(evanescent) 한다.

$$
\sigma = \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

$\sigma$ 가 실수가 되려면 $\beta > k_0 n_\text{clad}$ 이어야 하며, 이는 하한 조건과 일치한다.  
이 두 조건을 모두 만족하는 $\beta$ 구간에서만 완전 도파 모드가 존재한다.

### 복소 굴절율 (Complex refractive index) $N = n + ik $

평면파 파동함수를 $\psi(x) = Ae^{ikx}$ 라고 한다면  $k = Nk_0$ 이므로 

$$
\psi(x) = Ae^{i(k_0n+ik_0k)x} = Ae^{-k_0kx}e^{ik_0nx}
$$

즉, $e^{-k_0kx}$ 는 $x$ 에 따라서 진폭이 감쇄상수 $k_0k$로 급속도로 감쇠하는 **감쇠항**, $e^{ik_0nx}$ 는 파수 $k_0n$ 로 손실없이 진동하는 **진동항**으로 볼수있다.  이것을 위에서 서술한 코어내부의 횡방향 파수 $\kappa$ 와 클래드 내부의 횡방향 감쇠 상수 $\sigma$ 와 수학적으로 정합할지는 추후에 논의하겠다. 

---

## STEP 3 — 정규화된 횡방향 파수 $u$, $w$

수식을 다루기 편하게 코어 반경 $a$ 로 정규화한 무차원 파수를 도입한다.

$$
u = \kappa a = a \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$

$$
w = \sigma a = a \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

이 두 양은 다음과 같은 관계를 만족한다.

$$
u^2 + w^2 = a^2 k_0^2 (n_{\mathrm{core}}^2 - n_{\mathrm{clad}}^2) \equiv V^2
$$

여기서 $V$ 는 **수치 개구**(numerical aperture) 와 코어 반경을 합친 **V-파라미터** 혹은 **V-수**라 하며, 광섬유의 모드 수를 결정하는 핵심 양이다.  
$V < 2.405$ 이면 오직 기본 모드 $HE_{11}$만 전파 가능한 **단일 모드 섬유**가 된다.

---

## STEP 4 — Bessel 함수의 도함수 표기

경계 조건을 쓸 때 변형 Bessel 함수 $J_l(r)$, $K_l(r)$ 의 도함수가 자주 등장한다.  
이를 간결하게 표기하기 위해 다음 기호를 도입한다.

1. **제1종 베셀 함수 미분 $J_l'(r)$**
   $$J_d(l, r) = J_{l-1}(r) - \frac{l}{r} J_l(r)$$
   * 이 정의는 제1종 베셀 함수의 미분 공식과 같다.  
   * 즉, 시뮬레이션에서 $J_d(l, r) = J_l'(r)$ 성립한다.

2. **제2종 변형 베셀 함수 미분 $K_l'(r)$**
   $$K_d(l, r) = \frac{l}{r} K_l(r) - K_{l+1}(r)$$
   * 이 정의는 제2종 변형 베셀 함수의 미분 공식과 같다.  
   * 즉, 시뮬레이션에서 $K_d(l, r) = K_l'(r)$ 성립한다.

이러한 표기법을 도입하면 초월 방정식 내의 도함수들을 $J_l, K_l$ 및 차수가 하나 다른 $J_{l\pm1}, K_{l\pm1}$ 들의 선형 결합(Linear Combination) 형태로 간결하게 변환할 수 있어 수치 계산 및 코드 구현이 매우 편리해진다.

---

## STEP 5 — 비대칭 파라미터 $n_d$, $n_s$

코어와 클래드의 굴절률 차이를 두 가지 조합으로 정의한다.

$$
n_d = \frac{n_{\mathrm{core}}^2 - n_{\mathrm{clad}}^2}{2\,n_{\mathrm{core}}^2}
$$

$$
n_s = \frac{n_{\mathrm{core}}^2 + n_{\mathrm{clad}}^2}{2\,n_{\mathrm{core}}^2}
$$

- $n_d$ 는 코어와 클래드의 굴절률 **차이**를 나타내는 비대칭 파라미터로, 경계 조건에서 전계와 자계의 탄젠트 성분이 얼마나 다르게 결합하는지를 반영한다.  
  약 도파(weak guidance) 근사($n_\text{core} \approx n_\text{clad}$)에서는 $n_d \approx 0$ 이 되어 수식이 단순해진다.

- $n_s$ 는 굴절률의 **평균**에 가까운 값으로, 경계에서 경계 조건의 스케일을 조절하는 역할을 한다.

---

## STEP 6 — 보정 항 $R(\beta)$

$R(\beta)$ 는 벡터 파동 방정식에서 발생하는 **편광 보정 항**으로, EH 모드와 HE 모드를 구분하는 핵심 역할을 한다.

$$
R(\beta) = \sqrt{
  n_d^2 \left[\frac{K_l'(w)}{w\,K_l(w)}\right]^2
  + \left(\frac{l\,\beta}{n_{\mathrm{core}}\,k_0}\right)^2
  \left(\frac{1}{w^2} + \frac{1}{u^2}\right)^2
}
$$

- **첫 번째 항** $n_d^2[\cdots]^2$: 굴절률 불연속에 의한 편광 혼합 효과를 나타낸다. $n_d$ 가 클수록(굴절률 차이가 클수록) 이 항이 커진다.

- **두 번째 항**: 각 방위각 차수 $l$ 과 전파 상수 $\beta$ 에 의존한다. $l = 0$ 이면 이 항은 사라져 TE/TM 모드만 남는다.

$R(\beta)$ 는 항상 양의 실수이며, 이 값의 **부호**를 달리 취함으로써 EH 모드와 HE 모드의 초월 방정식을 통일적으로 기술할 수 있다.

---

## STEP 7 — 초월 방정식과 β의 결정

### EH 모드의 초월 방정식

$$
f_{\mathrm{EH}}(\beta) = \frac{J_{l+1}(u)}{u\,J_l(u)} - n_s\,\frac{K_l'(w)}{w\,K_l(w)} - \left(\frac{l}{u^2} - R(\beta)\right) = 0
$$

### HE 모드의 초월 방정식

$$
f_{\mathrm{HE}}(\beta) = \frac{J_{l+1}(u)}{u\,J_l(u)} - n_s\,\frac{K_l'(w)}{w\,K_l(w)} - \left(\frac{l}{u^2} + R(\beta)\right) = 0
$$

### 두 방정식의 차이

두 초월 방정식의 **유일한 차이**는 마지막 항의 $R(\beta)$ 의 부호다.

| 모드 | 마지막 항 | 의미 |
|------|----------|------|
| EH   | $\frac{l}{u^2} - R(\beta)$ | $R(\beta)$ 를 **빼는** 방향으로 편광 보정 |
| HE   | $\frac{l}{u^2} + R(\beta)$ | $R(\beta)$ 를 **더하는** 방향으로 편광 보정 |

이 부호 차이는 경계 조건에서 전계의 횡방향 성분이 코어와 클래드 경계에서 어떻게 이어지느냐에 따른 것으로, **전계가 주도하는 모드(EH)**와 **자계가 주도하는 모드(HE)**의 구조 차이를 반영한다.

### β의 수치적 결정 방법

1. 각 $l$ 에 대하여 $\beta \in (n_\text{clad} k_0,\; n_\text{core} k_0)$ 범위를 이산적으로 탐색한다.
2. $f_{\mathrm{EH}}(\beta)$ 또는 $f_{\mathrm{HE}}(\beta)$ 가 0이 되는 점을 **부호 변환(sign change)** 으로 찾는다.
3. 이분법(bisection) 또는 Brent's method 등으로 정밀하게 수렴시킨다.
4. 각 해 $\beta_{lm}$ (방위각 차수 $l$, 해당 방위각 차수에 대한 여러개의 해를 인덱싱한 반경 방향 차수 $m$)에 대응하는 모드 프로파일을 구성한다.

## STEP 8 — 모드 명명 규칙

| 모드 종류 | 설명 | 존재 조건 |
|----------|------|----------|
| $TE_{0m}$ | 전계가 횡단면 내에만 있음, $l=0$ | EH 방정식 $l=0$ 특수해 |
| $TM_{0m}$ | 자계가 횡단면 내에만 있음, $l=0$ | HE 방정식 $l=0$ 특수해 |
| $HE_{lm}$ | 자계 성분이 우세한 혼합 모드 | HE 방정식 $l \geq 1$ |
| $EH_{lm}$ | 전계 성분이 우세한 혼합 모드 | EH 방정식 $l \geq 1$ |

기본 모드 $HE_{11}$ 은 $l=1$ 의 HE 방정식에서 $m=1$ 에 해당하는 첫 번째 해이며, **차단 주파수가 0**이므로 어떤 V수에서도 항상 존재한다.

**TE 모드** 벡터파동방정식에서 $E_z = 0$ 인 모드<br>
**TM 모드** 벡터파동방정식에서 $H_z = 0$ 인 모드

## STEP 9 — 이상적인 상태에서 $\beta_{lm}$ 에 대한 각 모드의 공간적 분포 구하기

원통형 유전체 광섬유(Optical Fiber) 내부에서 전파되는 **하이브리드 모드의 전자기장**을 원통좌표계 ($\rho, \theta, z$)에서 계산하는 수식들로 구현

### 모드별 진폭계수 구하기 ($A, B, C, D$)
광섬유 전자기장의 $z$축 성분 진폭을 결정하는 상수들입니다. $l$값(방위각 차수)에 따라 수식이 달라집니다.

1. $l = 0$ 인 경우
    * **EH mode** (TM-like)
      $\displaystyle A = 1, B = 0, C = \frac{J_0(u)}{K_0(w)}, D = 0$
    * **HE mode** (TE-like)
      $\displaystyle A = 0, B = 1, C = 0, D = \frac{J_0(u)}{K_0(w)}$

2. $l \neq 0$ 인 경우 (하이브리드 모드)
    * $A = 1$
    * $B = \frac{il\beta}{\omega\mu_0} \left( \frac{1}{(w)^2} + \frac{1}{(u)^2} \right) \left[ \frac{J_l'(u)}{u J_l(u)} + \frac{K_l'(w)}{w K_l(w)} \right]^{-1}$
    * $C = \frac{J_l(u)}{K_l(w)}A$
    * $D = B \cdot C$

### 영역별 전자기장 수식
* **공통 위상항 (Phase term):** $\Phi = e^{i(l\theta - \beta z)}$
* 실제 시뮬레이션에서는 코어 영역에서의 $E_r, E_\theta$ 만 사용.

1.  코어 영역 ($\rho < a$) : 도파(Oscillating) 성분
$$
\begin{aligned}
E_r &= -\frac{i\beta}{\kappa^2} \left[ A \kappa J_l'(\kappa\rho) + \frac{i\omega\mu_0 l}{\beta\rho} B J_l(\kappa\rho) \right] \Phi \\
E_\theta &= -\frac{i\beta}{\kappa^2} \left[ \frac{il}{\rho} A J_l(\kappa\rho) - \frac{\omega\mu_0}{\beta} B \kappa J_l'(\kappa\rho) \right] \Phi \\
E_z &= A J_l(\kappa\rho) \Phi \\
\\
H_r &= -\frac{i\beta}{\kappa^2} \left[ B \kappa J_l'(\kappa\rho) - \frac{i\omega\epsilon_{\text{core}} l}{\beta\rho} A J_l(\kappa\rho) \right] \Phi \\
H_\theta &= -\frac{i\beta}{\kappa^2} \left[ \frac{il}{\rho} B J_l(\kappa\rho) + \frac{\omega\epsilon_{\text{core}}}{\beta} A \kappa J_l'(\kappa\rho) \right] \Phi \\
H_z &= B J_l(\kappa\rho) \Phi
\end{aligned}
$$

2.  클래딩 영역 ($\rho > a$) : 감쇠(Evanescent) 성분
$$
\begin{aligned}
E_r &= \frac{i\beta}{\sigma^2} \left[ C \sigma K_l'(\sigma\rho) + \frac{i\omega\mu_0 l}{\beta\rho} D K_l(\sigma\rho) \right] \Phi \\
E_\theta &= \frac{i\beta}{\sigma^2} \left[ \frac{il}{\rho} C K_l(\sigma\rho) - \frac{\omega\mu_0}{\beta} D \sigma K_l'(\sigma\rho) \right] \Phi \\
E_z &= C K_l(\sigma\rho) \Phi \\
\\
H_r &= \frac{i\beta}{\sigma^2} \left[ D \sigma K_l'(\sigma\rho) - \frac{i\omega\epsilon_{\text{clad}} l}{\beta\rho} C K_l(\sigma\rho) \right] \Phi \\
H_\theta &= \frac{i\beta}{\sigma^2} \left[ \frac{il}{\rho} D K_l(\sigma\rho) + \frac{\omega\epsilon_{\text{clad}}}{\beta} C \sigma K_l'(\sigma\rho) \right] \Phi \\
H_z &= D K_l(\sigma\rho) \Phi
\end{aligned}
$$
