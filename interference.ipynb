# Interference

## Interactive Simulation

Click 🚀 to run in Binder or run directly below.


```
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets

x = np.linspace(0, 10, 1000)

def plot_wave(phi=0.0):
    y = np.cos(x + phi)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y))
    fig.show()

widgets.interact(plot_wave, phi=(0, 6.28, 0.1))
```
