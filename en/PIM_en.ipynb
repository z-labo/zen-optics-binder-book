## STEP 1 — What Is the Propagation Constant β?

Inside an optical fiber, light propagates in specific spatial patterns called **modes**.  
Each mode has its own **propagation constant $\beta$**, which means that when the mode travels a distance $z$ along the axial direction (the z direction, also called the longitudinal direction), its phase is additionally shifted by $e^{i\beta z}$.

### Physical Meaning of β

In free space, the wave number is $k_0 = \omega/c$.  
In a homogeneous medium with refractive index $n$, it becomes $n k_0$.  
However, an optical fiber is a **waveguiding structure** made of a core (refractive index $n_\text{core}$) and a cladding (refractive index $n_\text{clad}$, with $n_\text{clad} < n_\text{core}$), so light behaves as if it propagates with an effective refractive index between the two.

$$
n_\text{clad} k_0 < \beta < n_\text{core} k_0
$$

Outside this range, the mode cannot exist.

- If $\beta \geq n_\text{core} k_0$, the entire core would effectively have the same phase, so no propagating solution exists.
- If $\beta \leq n_\text{clad} k_0$, the field does not decay in the cladding and becomes a **radiation mode**, with energy leaking away without bound.

As described later in STEP 7, only discrete values of $\beta$ can satisfy both boundary conditions simultaneously, and the corresponding transverse electromagnetic-field distributions are the **guided modes**.

---

## STEP 2 — Definition of Transverse Wave Numbers

Once the propagation constant $\beta$ is fixed, two wave numbers can be defined to describe the **transverse** behavior (oscillation in the x-y plane) inside the core and the cladding.

### Transverse Wave Number in the Core: $\kappa$

Inside the core, light **oscillates** in the transverse direction.

$$
\kappa = \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$

$k_0 n_\text{core}$ is the total wave number inside the core. After subtracting the axial component $\beta$, the remaining component is the transverse component (a Pythagorean vector decomposition). For $\kappa$ to be real, $\beta < k_0 n_\text{core}$ must hold, which matches the upper-bound condition above.

### Transverse Decay Constant in the Cladding: $\sigma$

In the cladding, light **decays exponentially** (evanescently) instead of oscillating.

$$
\sigma = \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

For $\sigma$ to be real, $\beta > k_0 n_\text{clad}$ must hold, which matches the lower-bound condition.  
Only within the range of $\beta$ satisfying both conditions can a fully guided mode exist.

### Complex Refractive Index $N = n + ik$

If the plane-wave wavefunction is written as $\psi(x) = Ae^{ikx}$, then $k = Nk_0$, so

$$
\psi(x) = Ae^{i(k_0n+ik_0k)x} = Ae^{-k_0kx}e^{ik_0nx}
$$

Thus, $e^{-k_0kx}$ can be regarded as a **decay term**, whose amplitude rapidly decays with $x$ with decay constant $k_0k$, while $e^{ik_0nx}$ can be regarded as an **oscillation term**, which oscillates without loss with wave number $k_0n$. How this mathematically matches the core transverse wave number $\kappa$ and the cladding transverse decay constant $\sigma$ described above will be discussed later.

---

## STEP 3 — Normalized Transverse Wave Numbers $u$ and $w$

To make the equations easier to handle, introduce dimensionless wave numbers normalized by the core radius $a$.

$$
u = \kappa a = a \sqrt{k_0^2 n_{\mathrm{core}}^2 - \beta^2}
$$

$$
w = \sigma a = a \sqrt{\beta^2 - k_0^2 n_{\mathrm{clad}}^2}
$$

These two quantities satisfy the following relation.

$$
u^2 + w^2 = a^2 k_0^2 (n_{\mathrm{core}}^2 - n_{\mathrm{clad}}^2) \equiv V^2
$$

Here, $V$ is the **V-parameter**, or **V-number**, combining the **numerical aperture** and the core radius. It is a key quantity that determines the number of modes in an optical fiber.  
If $V < 2.405$, only the fundamental mode $HE_{11}$ can propagate, and the fiber is a **single-mode fiber**.

---

## STEP 4 — Derivative Notation for Bessel Functions

When writing boundary conditions, derivatives of the Bessel functions $J_l(r)$ and $K_l(r)$ frequently appear.  
To write them compactly, introduce the following notation.

1. **Derivative of the Bessel function of the first kind $J_l'(r)$**
   * $J_d(l, r) = J_{l-1}(r) - \frac{l}{r} J_l(r)$
   * This definition is identical to the derivative formula for the Bessel function of the first kind.  
   * In the simulation, $J_d(l, r) = J_l'(r)$.

2. **Derivative of the modified Bessel function of the second kind $K_l'(r)$**
   * $K_d(l, r) = \frac{l}{r} K_l(r) - K_{l+1}(r)$
   * This definition is identical to the derivative formula for the modified Bessel function of the second kind.  
   * In the simulation, $K_d(l, r) = K_l'(r)$.

With this notation, the derivatives in the transcendental equations can be compactly converted into **linear combinations** of $J_l$, $K_l$, and the neighboring orders $J_{l\pm1}$ and $K_{l\pm1}$, making numerical calculation and code implementation much more convenient.

---

## STEP 5 — Asymmetry Parameters $n_d$ and $n_s$

Define the refractive-index difference between the core and cladding using two combinations.

$$
n_d = \frac{n_{\mathrm{core}}^2 - n_{\mathrm{clad}}^2}{2\,n_{\mathrm{core}}^2}
$$

$$
n_s = \frac{n_{\mathrm{core}}^2 + n_{\mathrm{clad}}^2}{2\,n_{\mathrm{core}}^2}
$$

- $n_d$ is an asymmetry parameter representing the refractive-index **difference** between the core and cladding. It reflects how differently the tangential components of the electric and magnetic fields are coupled at the boundary.  
  Under the weak-guidance approximation ($n_\text{core} \approx n_\text{clad}$), $n_d \approx 0$, which simplifies the equations.

- $n_s$ is close to the **average** of the refractive indices and controls the scale of the boundary conditions at the interface.

---

## STEP 6 — Correction Term $R(\beta)$

$R(\beta)$ is a **polarization correction term** arising from the vector wave equation, and it plays a central role in distinguishing EH modes from HE modes.

$$
R(\beta) = \sqrt{
  n_d^2 \left[\frac{K_l'(w)}{w\,K_l(w)}\right]^2
  + \left(\frac{l\,\beta}{n_{\mathrm{core}}\,k_0}\right)^2
  \left(\frac{1}{w^2} + \frac{1}{u^2}\right)^2
}
$$

- **First term** $n_d^2[\cdots]^2$: represents the polarization-mixing effect caused by the refractive-index discontinuity. The larger $n_d$ is (that is, the larger the refractive-index difference is), the larger this term becomes.

- **Second term**: depends on the azimuthal order $l$ and the propagation constant $\beta$. When $l = 0$, this term vanishes, leaving only TE/TM modes.

$R(\beta)$ is always a positive real number, and by choosing the **sign** of this quantity differently, the transcendental equations for EH and HE modes can be described in a unified way.

---

## STEP 7 — Transcendental Equations and Determination of β

### Transcendental Equation for EH Modes

$$
f_{\mathrm{EH}}(\beta) = \frac{J_{l+1}(u)}{u\,J_l(u)} - n_s\,\frac{K_l'(w)}{w\,K_l(w)} - \left(\frac{l}{u^2} - R(\beta)\right) = 0
$$

### Transcendental Equation for HE Modes

$$
f_{\mathrm{HE}}(\beta) = \frac{J_{l+1}(u)}{u\,J_l(u)} - n_s\,\frac{K_l'(w)}{w\,K_l(w)} - \left(\frac{l}{u^2} + R(\beta)\right) = 0
$$

### Difference Between the Two Equations

The **only difference** between the two transcendental equations is the sign of $R(\beta)$ in the final term.

| Mode | Final term | Meaning |
|------|------------|---------|
| EH   | $\frac{l}{u^2} - R(\beta)$ | Polarization correction in the direction of **subtracting** $R(\beta)$ |
| HE   | $\frac{l}{u^2} + R(\beta)$ | Polarization correction in the direction of **adding** $R(\beta)$ |

This sign difference comes from how the transverse electric-field components connect at the core-cladding boundary, and it reflects the structural difference between **electric-field-dominant modes (EH)** and **magnetic-field-dominant modes (HE)**.

### Numerical Procedure for Determining β

1. For each $l$, discretely search the range $\beta \in (n_\text{clad} k_0,\; n_\text{core} k_0)$.
2. Find where $f_{\mathrm{EH}}(\beta)$ or $f_{\mathrm{HE}}(\beta)$ becomes zero by detecting a **sign change**.
3. Refine the root using bisection, Brent's method, or another root-finding method.
4. Construct the mode profile corresponding to each solution $\beta_{lm}$, where $l$ is the azimuthal order and $m$ is the radial-order index that labels multiple solutions for that azimuthal order.

---

## STEP 8 — Mode Naming Rules

| Mode type | Description | Existence condition |
|-----------|-------------|---------------------|
| $TE_{0m}$ | Electric field exists only in the transverse plane, $l=0$ | Special solution of the EH equation for $l=0$ |
| $TM_{0m}$ | Magnetic field exists only in the transverse plane, $l=0$ | Special solution of the HE equation for $l=0$ |
| $HE_{lm}$ | Hybrid mode with dominant magnetic-field components | HE equation with $l \geq 1$ |
| $EH_{lm}$ | Hybrid mode with dominant electric-field components | EH equation with $l \geq 1$ |

The fundamental mode $HE_{11}$ is the first solution corresponding to $m=1$ in the HE equation for $l=1$, and because its **cutoff frequency is zero**, it always exists for any V-number.

**TE mode**: a mode with $E_z = 0$ in the vector wave equation<br>
**TM mode**: a mode with $H_z = 0$ in the vector wave equation

## STEP 9 — Finding the Spatial Distribution of Each Mode for $\beta_{lm}$ in the Ideal Case

Implemented using formulas that calculate the **electromagnetic fields of hybrid modes** propagating inside a cylindrical dielectric optical fiber in cylindrical coordinates ($\rho, \theta, z$).

### Finding Mode-Specific Amplitude Coefficients ($A, B, C, D$)
These are constants that determine the amplitudes of the z-axis components of the optical-fiber electromagnetic field. The formulas differ depending on the value of $l$ (azimuthal order).

1. For $l = 0$
    * **EH mode** (TM-like)
      $\displaystyle A = 1, B = 0, C = \frac{J_0(u)}{K_0(w)}, D = 0$
    * **HE mode** (TE-like)
      $\displaystyle A = 0, B = 1, C = 0, D = \frac{J_0(u)}{K_0(w)}$

2. For $l \neq 0$ (hybrid modes)
    * $A = 1$
    * $B = \frac{il\beta}{\omega\mu_0} \left( \frac{1}{(w)^2} + \frac{1}{(u)^2} \right) \left[ \frac{J_l'(u)}{u J_l(u)} + \frac{K_l'(w)}{w K_l(w)} \right]^{-1}$
    * $C = \frac{J_l(u)}{K_l(w)}A$
    * $D = B \cdot C$

### Electromagnetic-Field Formulas by Region
* **Common phase term:** $\Phi = e^{i(l\theta - \beta z)}$
* In the actual simulation, only $E_r$ and $E_\theta$ in the core region are used.

1. Core region ($\rho < a$): guided (oscillating) components

$$
\begin{aligned}

E_r &= -\frac{i\beta}{\kappa^2} \left[ A \kappa J_l'(\kappa\rho) + \frac{i\omega\mu_0 l}{\beta\rho} B J_l(\kappa\rho) \right] \Phi \\

E_\theta &= -\frac{i\beta}{\kappa^2} \left[ \frac{il}{\rho} A J_l(\kappa\rho) - \frac{\omega\mu_0}{\beta} B \kappa J_l'(\kappa\rho) \right] \Phi \\

E_z &= A J_l(\kappa\rho) \Phi \\

\\

H_r &= -\frac{i\beta}{\kappa^2} \left[ B \kappa J_l'(\kappa\rho) - \frac{i\omega\epsilon_{\text{core}} l}{\beta\rho} A J_l(\kappa\rho) \right] \Phi \\

H_\theta &= -\frac{i\beta}{\kappa^2} \left[ \frac{il}{\rho} B J_l(\kappa\rho) + \frac{\omega\epsilon_{\text{core}}}{\beta} A \kappa J_l'(\kappa\rho) \right] \Phi \\

H_z &= B J_l(\kappa\rho) \Phi \\

\\

\end{aligned}
$$

2. Cladding region ($\rho > a$): evanescent components

$$
\begin{aligned}

E_r &= \frac{i\beta}{\sigma^2} \left[ C \sigma K_l'(\sigma\rho) + \frac{i\omega\mu_0 l}{\beta\rho} D K_l(\sigma\rho) \right] \Phi \\

E_\theta &= \frac{i\beta}{\sigma^2} \left[ \frac{il}{\rho} C K_l(\sigma\rho) - \frac{\omega\mu_0}{\beta} D \sigma K_l'(\sigma\rho) \right] \Phi \\

E_z &= C K_l(\sigma\rho) \Phi \\

\\

H_r &= \frac{i\beta}{\sigma^2} \left[ D \sigma K_l'(\sigma\rho) - \frac{i\omega\epsilon_{\text{clad}} l}{\beta\rho} C K_l(\sigma\rho) \right] \Phi \\

H_\theta &= \frac{i\beta}{\sigma^2} \left[ \frac{il}{\rho} D K_l(\sigma\rho) + \frac{\omega\epsilon_{\text{clad}}}{\beta} C \sigma K_l'(\sigma\rho) \right] \Phi \\

H_z &= D K_l(\sigma\rho) \Phi \\

\\

\end{aligned}
$$
